9. <span style="font-size: 11pt; color: rgb(73, 68, 41); font-family: &quot;Cambria,Bold&quot;;">Group the sales data of each region on Year_of_Release and plot the same (It should&nbsp;</span> <span style="font-size: 11pt; color: rgb(73, 68, 41); font-family: &quot;Cambria,Bold&quot;;">show which region sales was at peak on which year of release)&nbsp;</span>

```python
#groupby() function is used to split the data into groups based on some criteria
#sum() which sums up the numbers in the Column 
Sales_year_Release =  df.groupby('Year_of_Release')['Global_Sales'].aggregate(np.mean)

Sales_NA = df.groupby('Year_of_Release')['NA_Sales'].aggregate(np.mean)
Sales_JP = df.groupby('Year_of_Release')['JP_Sales'].aggregate(np.mean)
Sales_EU = df.groupby('Year_of_Release')['EU_Sales'].aggregate(np.mean)
Sales_Other = df.groupby('Year_of_Release')['Other_Sales'].aggregate(np.mean)

Sales_All = df.groupby('Year_of_Release')[['Global_Sales','NA_Sales','EU_Sales','JP_Sales','Other_Sales']].aggregate(np.mean)
# plotting the line of all points
Sales_All.plot(figsize=(20,10))
# giving a title to my graph
plt.title('sales data of each region on Year_of_Release!')
```

10. <span style="font-size: 11pt; color: rgb(73, 68, 41); font-family: &quot;Cambria,Bold&quot;;">Group North American, Japan, European &amp; Other region sales data as per Genre and plot it (It should show highest of region according to game category , ex- NA_Sales is highest in Adventure category , plot kind should be bar)&nbsp;</span>

```python
#groupby() function is used to split the data into groups based on some criteria
#sum() which sums up the numbers in the Column 
group_region = df.groupby('Genre')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].aggregate(np.mean)
# plotting the line of all points
group_region.plot(kind='bar', figsize=(20,10))
```

11. <span style="font-size: 11pt; color: rgb(73, 68, 41); font-family: &quot;Cambria,Bold&quot;;">Group North American, Japan, European &amp; Other region sales data as per Platform and plot it (Should show sales as per "Platform genre" , plot kind should be bar )&nbsp;</span>
